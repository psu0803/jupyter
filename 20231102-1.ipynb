```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1, 5], [3, 4], [2, 3],
             [5, 3], [6, 2], [5, 1]])
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
y = kmeans.labels_
ycenter = kmeans.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.scatter(ycenter[:, 0], ycenter[:, 1], c=[0, 1], s = 200, cmap='viridis', edgecolor='k')

plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1, 5], [3, 4], [2, 3],
             [5, 3], [6, 2], [5, 1]])
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
x = np.array([[1, 4], [4, 2]])
y_pred = kmeans.predict(x)
print(y_pred)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap='magma', s = 200, edgecolor='k', marker = '^')

plt.show()
```
