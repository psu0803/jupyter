```python
import pandas as pd
data = pd.read_csv('acci.csv', encoding="cp949")
data
```

```python
x = data.iloc[:-1, 3:]
y = data.iloc[:-1, 2]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regr = DecisionTreeRegressor()
```

```python
regr.fit(x, y)
y_pred = regr.predict([[202, 11008, 3846, 6770, 392]])
print(y_pred)
```
